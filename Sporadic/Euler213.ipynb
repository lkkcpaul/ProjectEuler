```python
import numpy as np
```

```python
num_adj = np.ones((30,30),dtype=np.float128)*4
num_adj[0:-1,0] = 3
num_adj[0:-1,-1] = 3
num_adj[0,0:-1] = 3
num_adj[-1,0:-1] = 3
num_adj[0,0] = 2
num_adj[0,-1] = 2
num_adj[-1,0] = 2
num_adj[-1,-1] = 2
```

```python
def jump(input):
  output = np.zeros_like(input,dtype=np.float128)
  prob = input/num_adj
  output[1:, :] += prob[0:-1,:]
  output[0:-1, :] += prob[1:,:]
  output[:, 1:] += prob[:,0:-1]
  output[:,0:-1] += prob [:, 1:] 
  return output
```

```python
def jump50(board):
  for i in range(50):
    board = jump(board)
  return(board)
     
```

```python
prob_empty = np.ones((30,30))
for i in range(30):
  for j in range(30):
    board = np.zeros((30,30),dtype=np.float128)
    board[i,j]=1
    prob_of_one = 1 - jump50(board)
    prob_empty *= prob_of_one
```
