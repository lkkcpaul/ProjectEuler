Fast Solution by jp

```python
maxlevel=18
allways={}
ways=[1]*(maxlevel+1)

for i in range(maxlevel+1):
    ways[i]={}

ways[1][60]=60

for i in range(1,maxlevel+1):
    for j in range(1,math.floor(i/2)+1):
        for x in ways[j].keys():
            for y in ways[i-j].keys():
                ways[i][x+y]=1 
                ways[i][1.0*x*y/(x+y)]=1 
    for x in ways[i].keys():
        allways[round(x,9)]=x

print(len(allways.keys()))
```

My code using floats. Single capacitor has to be 60 to prevent rounding error.

```python
N = []
for i in range(18):
  N.append(set())
N[0]={60}

for i in range(17):
  for j in range(math.floor(i/2)+1):
#    temp1 = N[j]
#    temp2 = N[-1-j]
    for c1 in N[j]:
      for c2 in N[i-j]:
        N[i+1].add(c1+c2)
        N[i+1].add(1.0*c1*c2/(c1+c2))
  print (i+2, 'capacitors has ', len(N[i+1]), ' values')

Ways = set()
for i in range(18):
  for val in N[i]:
    Ways.add(round(val,10))
len(Ways)
```
