Convert each heap to binary. Add the numbers for each heap bitwise. If the result is 00000..., then the person to move loses. 

Because the heaps are n,2n,3n. For the bitwise sum to be 0000..., that means that n, when written in binary, must not have adjacent 1's. 

The question is then how to many ways to write a 31 digit binary number such that there are no adjacent 1s. '

```python
def ways(d):
  #d means the number of digits. And ways compute how many ways to write non-adjacent 1s. (Including 00000...)
  # Solve this recursively
  if d == 1:
    return 2
  if d == 2:
    return 3
  else:
    return ways(d-1) + ways(d-2)
```

```python
ways(30)
```
