```python
import math
class algebraic:
  ### this defines (a sqrt(n)+b)/c
  def __init__(self, n,a,b,c):
    self.a = a
    self.b = b
    self.c = c
    self.n = n
  
  def invert(self):
    newc = self.a**2 * self.n-self.b**2
    newa = self.c*self.a
    newb = -self.b*self.c
    dividor = math.gcd(math.gcd(newa,newb),newc)
    return(algebraic(self.n,newa//dividor,newb//dividor,newc//dividor))
 
  def minus(self, y):
    return(algebraic(self.n,self.a,self.b-y*self.c,self.c))

  def value(self):
    return ((self.a * math.sqrt(self.n) +self.b )/self.c)

  
  def write(self):
    print('(',self.a,' sqrt(',self.n,')+',self.b,')/',self.c )
```

```python
def period(N):
  x = algebraic(N,1,0,1)
  remainders = []
  i = 0
  while True:
    a = math.floor(x.value())
    r= x.minus(a)
    if r.value() in remainders:
      return(i - remainders.index(r.value()))
    remainders.append(r.value())
    i+=1
    x = r.invert()

count=0
print('processing')
for N in range(2,10001):
  if N%100 ==0:
    print(N,end=' ')
  if math.sqrt(N).is_integer():
    continue
  if period(N)%2 == 1:
    count+=1

print('\n',count)
```

This is the Pell's equation. Let n_i/d_i be the convergent series to the regular continued fraction for sqrt(D). 

Then the the first (n_i, d_i) pair that satisify x^2 - D y^2 = 1 is the minimal solution for the (x,y) pair.

We use the algorithm of p64 to find solutions. 



```python
import math
class algebraic:
  ### this defines (a sqrt(n)+b)/c
  def __init__(self, n,a,b,c):
    self.a = a
    self.b = b
    self.c = c
    self.n = n
  
  def invert(self):
    newc = self.a**2 * self.n-self.b**2
    newa = self.c*self.a
    newb = -self.b*self.c
    dividor = math.gcd(math.gcd(newa,newb),newc)
    return(algebraic(self.n,newa//dividor,newb//dividor,newc//dividor))
 
  def minus(self, y):
    return(algebraic(self.n,self.a,self.b-y*self.c,self.c))

  def value(self):
    return ((self.a * math.sqrt(self.n) +self.b )/self.c)

  
  def write(self):
    print('(',self.a,' sqrt(',self.n,')+',self.b,')/',self.c )
```

```python
from fractions import Fraction

def sol(N):
  z = algebraic(N,1,0,1)
  remainders = []
  i = 0

  alist = []
  while True:
    a = math.floor(z.value())
    
    frac = Fraction(1)
    if not alist:
      frac = Fraction(a)
    else:
      frac = Fraction(1,a)
      for a_i in alist[:0:-1]:
        frac = Fraction(1,a_i+frac)
      frac += Fraction(alist[0])
    alist.append(a)

    x = frac.numerator
    y = frac.denominator
    if x**2 - N*y**2 == 1:
      return(x)

    r= z.minus(a)
    if len(alist) >100:
      print('sol not found for D=', N)
      print('for x<=', x)
      return(0)
#    if r.value() in remainders:
#      print('not found')
#      return(0)
#    remainders.append(r.value())
    z = r.invert()
```

```python
maxminx = 0
theD = 0
for D in range(2,1001):
  if math.sqrt(D).is_integer():
    continue
  temp = sol(D)
#  print(D, temp)
  if temp> maxminx:
    maxminx=temp
    theD = D
print(theD)
```

```python
def num_rel_p(N, base, ind):
  count = 0
  j = ind
  while base*primes[j]<N:
    count = count + 1 + num_rel_p(N,base*primes[j],j)
    j += 1
  return count

```
