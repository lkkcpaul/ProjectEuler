```python
coins = [1,2,5,10,20,50,100,200]
def waystomake(tar, ind_max_coin):
  if ind_max_coin == 0:
    return 1
  ways = 0
  remain = tar
  while remain > 0:
    ways += waystomake(remain, ind_max_coin-1)
    remain -= coins[ind_max_coin]
  if remain == 0:
    ways += 1
  return ways
```

```python
import numpy as np
N=len(primes)
P = np.array(primes)
cumsum = np.cumsum(P)
```

```python
Pset = set(primes)
tar = 0
for c in range(21,N):
  if cumsum[c]>1000000:
    break
  for a in range(N-c):
    temp = cumsum[a+c]-cumsum[a]
    if temp>1000000:
      break
    if temp in Pset:
      tar=temp
      break
print(tar)
```
