```python
from collections import OrderedDict
import itertools
pool = list(itertools.combinations(range(10),4))
#operationorderpool = itertools.premutations(range(3))
operations = ['+','-','*','/']
operationpool = list(itertools.combinations_with_replacement(operations,3))

def operate(a,b,o):
  if o == '+':
    return a+b
  if o =='-':
    return a-b
  if o == '*':
    return a*b
  if o == '/':
    if b ==0:
      return float('inf')
    return a/b

def getnum(tup):
  output = set()
  for op in operationpool:
    expressions = list(itertools.permutations(list(op + tup)))

    for line in expressions:
      stack = []
      for c in line:
        if c in operations:
          if len(stack)<2:
            break
          else:
            stack.append(operate(stack.pop(-2),stack.pop(-1),c))
        else:
          stack.append(c)
      else:
        output.add(stack[0])

  return output
```

```python
maxN = 28
answer = (1,2,3,4)
for nums in pool:
  out = getnum(nums)
  n = 0 
  for i in range(1,100):
    if not i in out:
      n = i
      break
  if n>maxN:
    maxN = n
    answer = nums
```

**p94**

Area of triangle with side lengths a,b,c is
$$A = \sqrt{s(s-a)(s-b)(s-c)}$$ where $$s=\frac12(a+b+c)$$. Since the sides are a, a, a +/- 1, the formula simplifies to
$$A = \frac14 (a\pm 1)\sqrt{(3a\pm 1)(a\mp 1)}$$
a must be odd for A to be integral.

```python
import math
from decimal import *
N = 1000000000
S = 0
for a in range(3,N//3+1,2):
  X = (Decimal(3*a-1)*Decimal(a+1)/Decimal(4)).sqrt()
  Y = (Decimal(3*a+1)*Decimal(a-1)/Decimal(4)).sqrt()
  if X%1==0:
    S += 3*a-1
  if Y%1==0:
    S += 3*a+1

S
```

Not the most efficient but it does the job. Takes 10min to factorize all numbers.

```python
N = 1000000
```

```python
import math
primes = [2]
for i in range(3,N,2):
  j = 0 
  while primes[j]<= math.floor(math.sqrt(i)):
    if i % primes[j] == 0:
      break
    j += 1
  else:
    primes.append(i)
```

```python
primeset = set(primes)
```

```python
def SumOfFac(n):
  if n in primeset:
    return 1
  temp = n
  factors = []
  powers = []
  j = 0
  while temp != 1:
    p = primes[j]
    q = 0
    while temp%p == 0:
      temp = temp/p
      q+=1
    if q:
      factors.append(p)
      powers.append(q)
    j += 1
  S = 1
  for i in range(len(factors)):
    p = factors[i]
    q = powers[i]
    S*= (p**(q+1)-1)//(p-1)
  
  return(S-n)
```

```python
import numpy as np
TableSumFac = np.zeros(N+1)
for i in range(2,N+1):
  TableSumFac[i] = SumOfFac(i)
```

```python
TableSumFac = TableSumFac.astype(int)
```

```python
LongestChain = [6]
Len = 1
for i in range(2,N+1):

  is_chain = True
  cur = i
  sf = TableSumFac[i]
  chain = []
  while sf != i:
    if sf==0 or sf==1 or sf>N or (sf in chain and sf!=i):
      is_chain = False
      break
    chain.append(cur)
    cur = sf
    sf = TableSumFac[cur]
  if is_chain:
    if len(chain) > Len:
      Len = len(chain)
      LongestChain = chain.copy()
```

```python
def findblock(x,y):
  return((y//3) + (x//3)*3)
def blockcoords(z):
  temp = []
  xs = (z//3)*3
  ys = (z%3)*3
  for i in range(3):
    for j in range(3):
      temp.append((xs+i,ys+j))
  return(temp)
```

```python
def eliminate(x,y,n,poss):
  temp = np.copy(poss)
  temp[:,y,n-1] = np.zeros(9)
  temp[x,:,n-1] = np.zeros(9)
  for coords in blockcoords(findblock(x,y)):
    xb, yb = coords
    temp[xb,yb,n-1] = 0 
  temp[x,y,:] = np.zeros(9)
  return(temp) 
```

```python
def consistent(board,poss):
  for x in range(9):
    for y in range(9):
      if board[x][y] == '0' and np.sum(poss,axis=2)[x][y] == 0:
        return False
  return True
```

```python
# Find if there are any forced moves. If so, then check that move is valid and make next move
# Else iterate through the possibilities for the square with least number of possibilities.
# For each move, check if the move is valid and go on.
# If there is an 

def fill(board, poss): 
  if not consistent(board,poss):
    return False

  for rows in board:
    if '0' in rows:
      break
  else: #The board is filled
    print(board)
    upl = int(board[0][0:3])
    return upl

  for x in range(9):
    for y in range(9):
      if board[x][y]=='0' and np.sum(poss,axis=2)[x][y] == 1:
        digit = np.where(poss[x,y,:]==1)[0][0] + 1
        newboard = board.copy()
        newboard[x] = board[x][:y]+str(digit)+board[x][y+1:]
        newposs = eliminate(x,y,digit,poss)
        return fill(newboard,newposs)
  for x in range(9):
    for y in range(9):
      if board[x][y] !='0':
        continue
      for z in range(9):
        if poss[x,y,z]==0:
          continue
        digit = z+1
        newboard = board.copy()
        newboard[x] = board[x][:y]+str(digit)+board[x][y+1:]
        newposs = eliminate (x,y,digit,poss)
        valid = fill(newboard,newposs)
        if valid:
          return valid
      return False


```

```python
def startposs(board):
  poss = np.ones((9,9,9))
  for x in range(9):
    for y in range(9):
      if board[x][y] == '0':
        continue
      poss = eliminate(x,y,int(board[x][y]),poss)
  return poss
```

```python
S = 0
for g in games:
  poss = startposs(g)
  upl = fill(g,poss)
  S+=upl
print(S)
```

```python
anagrams = dict()
for word in data:
  temp = ''.join(sorted([c for c in word]))
  if temp in anagrams.keys():
    anagrams[temp].add(word)
  else:
    anagrams[temp] = {word}
  
```

```python
anapairs = []
for pair in anagrams.values():
  if len(pair)>1:
    anapairs.append(pair)
```

```python
import math
import itertools
largest = 0

for pair in anapairs:
  temp = list(pair)
  word1 = temp[0]
  word2 = temp[1]
  #word3 = pairs[2]

  structure1 = []
  used_char = []
  i = 0
  for c in word1:
    if c in used_char:
      structure1.append(used_char.index(c))
    else:
      structure1.append(i)
      i += 1
      used_char.append(c)

  structure2 = []
  for c in word2:
    structure2.append(used_char.index(c))

  num_distinct_digits = len(used_char)

  pool = list(itertools.permutations([str(i) for i in range(10)],num_distinct_digits))

  for num_map in pool:
    num1 = int(''.join([num_map[i] for i in structure1]))
    if not math.sqrt(num1).is_integer():
      continue
    num2 = int(''.join([num_map[i] for i in structure2]))
    if not math.sqrt(num2).is_integer():    
      continue
    if len(str(num1))!=len(str(num2)):
      continue
#    print(word1,word2,num1,num2)
    if max(num1,num2)>largest:
      largest = max(num1,num2)
```

Let Tn be the nth triangle number. 

If Tc *2 = Td, then T_{2c+d+1} = (c+d+1)^2 which means the triangle is a perfect square.

Then T_{3c+2d+2} *2 = T_{4c+3d+3}.

```python
c = 2
d = 3
tri = lambda n : n(n+1)//2
n = [c]
while(d<10**12):
  new_c = 3*c+2*d+2
  d = 4*c+3*d+3
  c = new_c
  assert(tri(c)*2==tri(d))
  n.append(c)
print(n[-1]+1)
```
